## Faktafrågor
1. Vad är *prompt engineering*? 

Justering av promptning för att LLM:s ska kunna tolka en fråga bättre.

2. Vad är *RAG*?

Reretievla augmented generation, funktioner som tar ort onödig data föra tt underlätta för modellen så att de fokuserar på rätt ord/meningar.

3. Vad är *chunking* och *embeddings*? 

Uppdelning av data som texter för att det ska bli lätttare för modellen. Embedding converterar text till vektor-stationer och gör semnatiska sökningar i embeddings. Cosinuslikheten beskrivber lekheten mellan vektorer.

5. Din kollega frågar dig vad ELIZA och Turingtestet är. Läs nedanstående två länkar innan du besvarar frågan.
* [https://en.wikipedia.org/wiki/ELIZA](https://en.wikipedia.org/wiki/ELIZA)
* [https://sv.wikipedia.org/wiki/Turingtestet](https://sv.wikipedia.org/wiki/Turingtestet)


Turingtestet fungerar så att man ställer samma fråga till en människa som man gör till en chattbot tillexempel. Om man får liknande svar eller samma så evalueras det som godkänt. ELIZA var den första chatbotten som kom ut på 1960-talet, man använde det för psykologiskt syfte.
