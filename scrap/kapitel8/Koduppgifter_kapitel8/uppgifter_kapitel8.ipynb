
## Faktafrågor
1. Vad krävs det för att ett neuralt nätverk ska klassas som CNN? 
Ifall det har ett convolution layers. (Oftast används också max pooling layers.)


2. Inom vilket tillämpningsområde är CNN generellt sett en väldigt kraftfull modell? 
Inom bildanalys, CNN är väldigt kraftfullt när man vill läsa av bilder och videor.


3. Vad menas med RGB? 
Det är 3 färg kanaler Red, Green och Blue (RGB)


4. Vad är *data augmentation*? 
Man justerar datan för att träna modellen på variation. tex, rotation eller spegelvändning av en bild.
