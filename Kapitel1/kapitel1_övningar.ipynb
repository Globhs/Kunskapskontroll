## 3. Förklara följande

### A. Vad är syftet med att dela upp data i träningsdata, valideringsdata och testdata?

Syftet är att träna olika modeller på träningsdatan, använda valideringsdatan för att välja och justera den modell som fungerar bäst, och sedan utvärdera den slutliga modellen på testdatan för att mäta dess generaliseringsförmåga.

### B. Vad är k-delad korsvalidering för något?

Det är när man delar upp datan i k-delar som sedan körs k-gånger i modellen, efter det så tar man medelvärdet av alla resultat för att få ett mer pålitgligt svar.

### C. Vad är root mean squared error (RMSE) för något?

RMSE är ett utvärderingsmått för regressionsproblem.
"Beräknar kvadratroten av medelvärdet av kvadrerade fel mellan de predikterade värdena och de faktiska värdena."

### D. Vad är en hyperparameter för något? Vad är en parameter för något?

Hyperparameter styr "hur modellen tränas" (t.ex. lärhastighet, antal lager, maxdjup), medan parameter är "vad modellen lärt sig" under träningen (t.ex. vikter och bias).

### E.

GridSearch är ett automatiserat sätt att välja hyperparametrar 
1. Du anger ett grid med möjliga värden för varje hyperparameter.
2. GridSearch testar alla unika kombinationer av dessa värden.
3. För varje kombination används ofta k-delad korsvalidering för att mäta prestanda.
4. Den kombination som ger bäst resultat väljs.


Det är därför den heter "grid" "search" då den letar efter den bästa kombinationen av punkter i ett rutnät.


Refit=True betyder att modellen tränas igen på hela träningsdatan med den hyperparameterkombination som GridSearch har hittat ger bäst prestanda.

### F. 

Det handlar om att ML-modeller behöver numerisk indata men vi vet också att att kattegorisa data är viktigt då datan inte har en speciell rangordning alltså "nominal data". För att omvandla denna data till numeriska värden använder vi oss utav dummy-variable-encoding och one-hot-encoding. Om det är ordinal data (data som har en speciell rangordning) så kan vi använda oss av en metod som heter ordinal-encoding. Ordinal och nominal data beror helt på hur vi tolkar datan i olika sammanhang. 

### G. Vad är feature engineering för något?

Processen att välja, skapa och omvandla variabler (features) i datan för att göra den bättre för modellen. När vi skapar modeller med en dålig variablselektion så får vi en opålitlig modell som leder till dålig prestanda och predektioner. "shit in shit out"

### H. Vad menas med principle of parsimony?

När flera modeller presterar ungefär lika bra bör man välja den enklaste modellen. För att inte riskera att få en överanpassad modell.

### 4. Vad menas med att "en modell är en förenkling av verkligheten?"

En modell är aldrig en exakt kopia av verkligheten. Den är en förenklad representation som fokuserar på de viktigaste aspekterna för att förstå, förklara eller förutsäga något.

### 5. Vad menas med att en modell är "överanpassad" eller overfitted på engelska?

När en modell är överanpassad så betyder det att du kommer få olika svar beroende på datan, alltså att modellen inte kan generalisera på ett bra sett därmed är den överanpassad.

### 6. Högre är bättre i scikit-learn scoring, vad innebär det?

I scikit-learn är det en allmän princip att "högre värden är bättre" i samband med att score beräknas. MSE och NMSE.

### 7. Vad är tvärsnittsdata, tidsseriedata och paneldata? Exemplifiera när respektive datakategori kan uppstå.

Tvärsnittsdata är sådan data där observationer kring olika individer förs tillexempel, hur inkomst och ålder håller för olika personer.

Tidsseriedata är sådan data där observationer över tid samlas in för en individ tillexempel, hur inkomst och ålder håller ihop för en enstaka individ under x antal år.

Paneldata är sådan data där observationer för olika individer, ett exempel på detta är hur inkomst och ålder håller ihop för tre olika personer under x antal år.

### 8. Ge några exempel på verkliga tillämpningsområden inom ML. Sök gärna på nätet för besvara frågan.

Diagnostik och riskbedömning där ML används för att analysera patientdata och röntgenbilder för att identifiera sjukdomar som cancer eller hjärtproblem.

Användardata från socialmedier där ML används för att analysera användardatan för att kunna rekomendera fler videos som kommer leda till att användaren använder tjänsten mer.

### 9. Generellt sett häller det att högre är bättre i scikit-learn scoring, därför används exempelvis scoring='neg_mean_squared_error*. Förklara logiken bakom detta, det vill säga att vi använder "negative" mean squared error.

Eftersom scikit-learn standardiserar scoring så att “högre är bättre” används neg_mean_squared_error för regressionsproblem. Då blir lägre fel omvänt till ett högre värde, vilket gör det lättare att rangordna modeller.
